```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

df.set_index('Order Date', inplace=True)
df2 = df.resample('D').sum()
```

```python
train_data = df[:-300]  # Example: use all but the last 30 days for training
test_data = df[-300:]   # Example: use the last 30 days for testing
```

```python
p, d, q = 1, 2, 1  # Example non-seasonal orders
P, D, Q, s = 1, 0, 1, 12  # Example seasonal orders (assuming monthly seasonality)


model = SARIMAX(train_data['Sales'], order=(p, d, q), seasonal_order=(P, D, Q, s))
results = model.fit()
predictions = results.get_forecast(steps=len(test_data)).predicted_mean
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test_data['Sales'], predictions)
print("MSE:", mse)
```

```python
rmse = mean_squared_error(test_data['Sales'], predictions, squared=False)
print("RMSE:", rmse)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(test_data['Sales'], predictions)
print("MAE:", mae)
```
