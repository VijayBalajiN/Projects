```python
import pandas as pd
df['Order Date'] = pd.to_datetime(df['Order Date'],format='%d/%m/%Y')
df["Ship Date"]=pd.to_datetime(df["Ship Date"],format='%d/%m/%Y')
```

```python
df["num_days"]=(df["Ship Date"]-df["Order Date"]).dt.days
```

```python
df
```

```python
df.drop(columns="Ship Date",inplace=True)
```

```python
df
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

df.set_index('Order Date', inplace=True)
df2 = df.resample('D').sum()  # Aggregate sales per day

# Explore time series (plot, ACF, PACF)

# Determine SARIMA parameters (experiment based on analysis)
p, d, q = 1, 1, 1  # Example non-seasonal orders
P, D, Q, s = 1, 0, 1, 12  # Example seasonal orders (assuming monthly seasonality)

# Fit SARIMA model
model = SARIMAX(df2['Sales'], order=(p, d, q), seasonal_order=(P, D, Q, s))
results = model.fit()
```

```python
train_data = df[:-300]  # Example: use all but the last 30 days for training
test_data = df[-300:]   # Example: use the last 30 days for testing
```

```python
p, d, q = 1, 2, 1  # Example non-seasonal orders
P, D, Q, s = 1, 0, 1, 12  # Example seasonal orders (assuming monthly seasonality)

# Fit SARIMA model
model = SARIMAX(train_data['Sales'], order=(p, d, q), seasonal_order=(P, D, Q, s))
results = model.fit()
predictions = results.get_forecast(steps=len(test_data)).predicted_mean
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test_data['Sales'], predictions)
print("MSE:", mse)
```

```python
rmse = mean_squared_error(test_data['Sales'], predictions, squared=False)
print("RMSE:", rmse)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(test_data['Sales'], predictions)
print("MAE:", mae)
```
